Frame the message

```python
message = "What is the date today?"

messages = [{"role": "user", "content": message}]
```

Connect to chatgpt using openapi library

```python
from openai import OpenAI

openai = OpenAI(api_key=OPENAI_API_KEY)

response = openai.chat.completions.create(model="gpt-5-nano", messages=messages)
response.choices[0].message.content
```

Connect to Google Gemini library. It shows the build date.

```python
from google import genai

gemini = genai.Client(api_key=GEMINI_API_KEY)

response = gemini.models.generate_content(model="gemini-2.5-flash", contents=message)
print(response.text)
```

Calling Anthropic API. Usually returns a message that it is not connected to internet and that's why it can't tell the date.

```python
import anthropic

client = anthropic.Anthropic(api_key=ANTHROPIC_API_KEY)

message = client.messages.create(model="claude-sonnet-4-5",  messages=messages)
print(message.content)
```
