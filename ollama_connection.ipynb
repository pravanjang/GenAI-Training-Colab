start_ollama method is required only on colab.

```python
import subprocess
import threading

def start_ollama():
    subprocess.run(["ollama", "serve"])

ollama_thread = threading.Thread(target=start_ollama)
ollama_thread.start()
```

Wait for sometime (few seconds) and download a model

```python
!ollama pull gpt-oss
```

Then call the ollama server

```python
from openai import OpenAI

ollama = OpenAI(base_url="http://localhost:11434/v1", api_key='ollama')

response = ollama.chat.completions.create(model="gpt-oss", messages=[{"role": "user", "content": "What is the date today?"}])

print(response.choices[0].message.content)
```
